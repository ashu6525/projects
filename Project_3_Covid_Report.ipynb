```python
data.groupby("Region")['Confirmed', 'Deaths', 'Recovered'].sum()
```

```python
data.groupby("Region").Confirmed.sum()
```

Q.4:- In which region, maximum number of Deaths cases were recorded?

```python
data.groupby("Region").Deaths.sum().sort_values()
```
